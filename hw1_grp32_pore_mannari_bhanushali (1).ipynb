```python
#referred from https://stackoverflow.com/questions/43798377/one-hot-encode-categorical-variables-and-scale-continuous-ones-simultaneouely
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LinearRegression,Lasso,ElasticNet,Ridge
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd

d = pd.read_csv('data.csv')
```

```python
numeric_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors','sqft_above', 'sqft_basement','yr_built', 'yr_renovated']
```

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])
```

```python
categorical_features = ['waterfront', 'view','condition','city']
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
preprocessing = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])
```

```python
X = d.drop(['date','price'], axis=1)
y = d['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
linearScore = Pipeline(steps=[('preprocessor', preprocessing),
                      ('classifier', LinearRegression())])

linearScore.fit(X_train, y_train)
linearScore.score(X_train, y_train)
```

```python
LassoScore = Pipeline(steps=[('preprocessor', preprocessing),
                      ('classifier', Lasso())])

LassoScore.fit(X_train, y_train)
LassoScore.score(X_train, y_train)
```

```python
EScore = Pipeline(steps=[('preprocessor', preprocessing),
                      ('classifier', ElasticNet())])
EScore.fit(X_train, y_train)
EScore.score(X_train, y_train)
```

```python
RidgeScore = Pipeline(steps=[('preprocessor', preprocessing),
                      ('classifier', Ridge())])
RidgeScore.fit(X_train, y_train)
RidgeScore.score(X_train, y_train)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import NearestNeighbors
from sklearn.model_selection import train_test_split
```

```python
numeric_features=['duration','credit_amount','installment_commitment','age','residence_since','existing_credits','num_dependents']
```

```python
category=['checking_status', 'credit_history', 'purpose','savings_status','employment','personal_status','other_parties','property_magnitude','other_payment_plans','housing','job','own_telephone','foreign_worker']
```

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])
```

```python
preprocessing = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, category)])
```

```python
y1= frame['credit_amount']
x1 = frame
X_train, X_test, y_train, y_test = train_test_split(x1, y1 , test_size=0.2)
logisticrScore = Pipeline(steps=[('preprocessor', preprocessing),
                      ('classifier', LogisticRegression())])
```

```python
logisticrScore.fit(X_train, y_train)
logisticrScore.score(X_train, y_train)
```

```python
SVCScore = Pipeline(steps=[('preprocessor', preprocessing),
                      ('classifier', SVC())])

SVCScore.fit(X_train, y_train)
SVCScore.score(X_train, y_train)
```
